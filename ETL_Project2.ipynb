```python
import pandas as pd
```

```python
saas_df = pd.read_csv("latka_web_data.csv")
saas_df.head()
```

```python
saas_df[['financials.arr_max', 'financials.arr_min', 'financials.revenue_2016', 'financials.revenue_2017', 'financials.num_customers_max', \
         'financials.num_customers_min']]
```

```python
def clean_url(url):
    if url[:8] == "https://":
        url = url[8:]
        
    elif url[:7] == "http://": 
        url = url[7:]
    
    if url[:4] == 'www.':
        url = url[4:]
        
    if url[-1] == '/':
        url = url[:-1]
    
    return url
```

```python
saas_df['url'] = saas_df['url'].apply(lambda a: clean_url(a))
```

```python
martech_df = pd.read_csv("martech2017.csv")
martech_df.head()
```

```python
martech_df.rename(columns = {"Website Domain":"url"}, inplace = True)
```

```python
martech_df[['Category', 'url']].head()
```

```python
maetech_url = martech_df[['Company', 'url', "Category", "Sub-Category"]]
saas_url = saas_df[["url", "industry", "sub_category", 'financials.arr_max', 'financials.arr_min', 'financials.revenue_2016', 'financials.revenue_2017', 'financials.num_customers_max', 'financials.num_customers_min']]
```

```python
saas_url.head() 
```

```python
df_merged = pd.merge(maetech_url, saas_url, on='url')
```

```python
df_merged = df_merged.drop_duplicates(subset="url").reset_index(drop=True)
```
